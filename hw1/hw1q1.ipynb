#**Problem 1**

Find three simple functions from R^2 to R and suitable points that are:

- a local minimum
- a local maximum
- neither a minimum nor maximum, but the gradient is zero at this point.

Plot these functions using 3D plots as in the notebooks for creating density and contour plots and three dimensional plots. Indicate clearly the points. Use 2d and 3d contour plots.

```
# Import libraries
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```
# Define function
def f(x, y):
  return x * (np.e ** (-1 * x ** 2 - (y **2)))
x = np.linspace(-2, 2)
y = np.linspace(-2, 2)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Local Min and Max definitions
local_min = (-1 / np.sqrt(2), 0, (-1 / np.sqrt(2 * np.e)))
local_max = (1 / np.sqrt(2), 0, (1 / np.sqrt(2 * np.e)))
```

```
# Graph function
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_zlim(-0.5, 0.5)
ax.plot_surface(X, Y, Z,)
ax.view_init(10, 75)

## Plot the:
# Local Minimum
ax.text(local_min[0],local_min[1],local_min[2],"  Local Minimum")
ax.plot([local_min[0]],[local_min[1]],[local_min[2]], markerfacecolor='black', markersize=6, marker='o')
# Local Maximum
ax.text(local_max[0],local_max[1],local_max[2],"  Local Maximum")
ax.plot([local_max[0]],[local_max[1]],[local_max[2]], markerfacecolor='black', markersize=6, marker='o')
fig.show()
```

```
# Graph 3D contour
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_zlim(-0.5, 0.5)
ax.contour3D(X, Y, Z, 50)
ax.view_init(10, 75)

## Plot the:
# Local Minimum
ax.text(local_min[0],local_min[1],local_min[2],"  Local Minimum")
ax.plot([local_min[0]],[local_min[1]],[local_min[2]], markerfacecolor='black', markersize=6, marker='o')
# Local Maximum
ax.text(local_max[0],local_max[1],local_max[2],"  Local Maximum")
ax.plot([local_max[0]],[local_max[1]],[local_max[2]], markerfacecolor='black', markersize=6, marker='o')
fig.show()
```

```
# Graph 2D contour
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('2D Contour')

## Plot the:
# Local Minimum
ax.text(local_min[0], local_min[1], "  Local Minimum")
ax.plot([local_min[0]],[local_min[1]], markerfacecolor='black', markersize=6, marker='o')
# Local Maximum
ax.text(local_max[0], local_max[1], "  Local Maximum")
ax.plot([local_max[0]],[local_max[1]], markerfacecolor='black', markersize=6, marker='o')
fig.show()
```

```
# Define function
def g(x, y):
  return x**3 - (3 * x * (y**2))
x2 = np.linspace(-0.5, 0.5)
y2 = np.linspace(-0.5, 0.5)

X2, Y2 = np.meshgrid(x2, y2)
Z2 = g(X2, Y2)

# Saddle point lies on 
saddle = (0, 0, 0)
```

```
# Graph the function
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_zlim(-0.25, 0.25)
ax.plot_surface(X2, Y2, Z2)
ax.view_init(10, 45)

# Plot the Saddle Point
ax.text(saddle[0], saddle[1], saddle[2], "  Saddle Point")
ax.plot([saddle[0]],[saddle[1]],[saddle[2]], markerfacecolor='black', markersize=6, marker='o')
fig.show()
```

```
# Graph 3D contour
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_zlim(-0.25, 0.25)
ax.contour3D(X2, Y2, Z2, 50)
ax.view_init(10, 45)

# Plot the Saddle Point
ax.text(saddle[0], saddle[1], saddle[2], "  Saddle Point")
ax.plot([saddle[0]],[saddle[1]],[saddle[2]], markerfacecolor='black', markersize=6, marker='o')
fig.show()
```

```
# Graph 2D contour
fig, ax = plt.subplots()
CS = ax.contour(X2, Y2, Z2)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('2D Contour')

# Plot the Saddle Point
ax.text(saddle[0], saddle[1],"  Saddle Point")
ax.plot([saddle[0]],[saddle[1]], markerfacecolor='black', markersize=6, marker='o')
fig.show()
```
